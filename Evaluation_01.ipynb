```python
#problem statement
#Predict the Yearly Income of individuals based on their
#personal and demographic features such as FirstName, MiddleName,
#LastName, AddressLine1, City, StateProvinceName, CountryRegionName, PostalCode, PhoneNumber, BirthDate,
#Education, Occupation, Gender, MaritalStatus, HomeOwnerFlag, NumberCarsOwned, NumberChildrenAtHome, TotalChildren.
```

```python
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

```python
#Predictor Attribute
X.head()
```

```python
#Target Attribute
Y.head()
```

```python
#: Split the Data into Train and Test
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
#Training Data
print("X_test", X_train.shape)
print("X_test", y_train.shape)
```

```python
#Testing Data
print("X_test", X_test.shape)
print("X_test", y_test.shape)
```

```python

#label encodinggg
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['StateProvinceName'] = label_encoder.fit_transform(df['StateProvinceName'].values)
df.head()
```

```python
df['CountryRegionName'] = label_encoder.fit_transform(df['CountryRegionName'].values)
df.head()
```

```python
df['Occupation'] = label_encoder.fit_transform(df['Occupation'].values)
df.head()
```

```python
df['Gender'] = label_encoder.fit_transform(df['Gender'].values)
df.head()
```

```python
df['MaritalStatus'] = label_encoder.fit_transform(df['MaritalStatus'].values)
df.head()
```

```python
df.drop('FirstName', axis = 1, inplace = True)
df.drop('MiddleName', axis = 1, inplace = True)
df.drop('AddressLine1', axis = 1, inplace = True)
df.drop('LastName', axis = 1, inplace = True)
df.drop('PostalCode', axis = 1, inplace = True)
df.drop('PhoneNumber', axis = 1, inplace = True)
df.drop('BirthDate', axis = 1, inplace = True)
df.drop('City', axis = 1, inplace = True)
```

```python
df.head()
```

```python
df['Education'] = label_encoder.fit_transform(df['Education'].values)
df.head()
```

```python
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
SS = StandardScaler()
X_train = SS.fit_transform(X_train)
X_test = SS.transform(X_test)

pd.DataFrame(X_test).head()  # Same Scale
```

MODEL

```python
#get the input shape
no = X_train.shape[1]
```

```python
no
```

```python
import tensorflow as tf

#set random seed
tf.random.set_seed(42)
#create a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(150,input_shape=(no,),activation='relu'),
    tf.keras.layers.Dense(150,activation = 'relu'),
    tf.keras.layers.Dense(1)#output node
])
model.compile(loss=tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])
```

```python
model.summary()
```

```python
model.fit(tf.expand_dims(X_train,axis=-1),y_train,epochs=100)
```

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(model , show_shapes=True)
```

```python
model.evaluate(X_train,y_train)
```

```python
model.evaluate(X_test,y_test)
```

```python
#make predictions
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
test_mse_score, test_mae_score = model.evaluate(X_test,y_test)
```

```python
y_test.shape,y_pred.shape
```

```python
mae = tf.metrics.mean_absolute_error(y_true=y_test,
                                     y_pred=y_pred.squeeze())
mae
```

```python
mse = tf.metrics.mean_squared_error(y_true=y_test,
                                    y_pred=y_pred.squeeze())
mse
```

```python
#MODEL-2
#set random seed
tf.random.set_seed(42)
#create a model
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(150,input_shape=(no,),activation='sigmoid'),
    tf.keras.layers.Dense(150,activation = 'sigmoid'),
    tf.keras.layers.Dense(150,activation = 'sigmoid'),
    tf.keras.layers.Dense(150,activation = 'sigmoid'),
    tf.keras.layers.Dense(1)#output node
])
model_2.compile(loss=tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(),
              metrics = ["mae"])
```

```python
model_2.summary()
```

```python
model_2.fit(tf.expand_dims(X_train,axis=-1),y_train,epochs=100)
```

```python
plot_model(model_2 , show_shapes=True)
```

```python
#make predictions
y_pred2 = model_2.predict(X_test)
y_pred2
```

```python
mae2 = tf.metrics.mean_absolute_error(y_true=y_test,
                                     y_pred=y_pred2.squeeze())
mae2
```

```python
mse2 = tf.metrics.mean_squared_error(y_true=y_test,
                                     y_pred=y_pred2.squeeze())
mse2
```

```python
#Comparing the results of all 3 models
model_results = [["model_1",mae.numpy(),mse.numpy()],
                 ["model_2",mae2.numpy(),mse2.numpy()],
]
```

```python
all_results = pd.DataFrame(model_results, columns=["model", "mae", "mse"])
all_results
```
